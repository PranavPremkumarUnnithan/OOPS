Theory Questions

1. What is Object-Oriented Programming (OOP)?

  - Object-Oriented Programming (OOP) is a way of writing code where we organize data and actions into objects.
  - These objects are like real-world things; they have properties and actions.
  - OOP helps make code easier to manage, reuse, and understand.


2. What is a class in OOP?

  - A class in OOP is like a blueprint for creating objects. It defines the properties and actions that the objects will have.


3. What is an object in OOP?

  - An object in OOP is a thing created from a class.
  - It has the properties and actions defined by the class.

4.  What is the difference between abstraction and encapsulation ?

  - Abstraction hides complex details and shows only the important parts to the user.
  - It helps to focus on what something does, not how it works.
  - Encapsulation hides the data inside an object and controls how it is accessed or changed.
  - It keeps data safe and organized by using private variables and public methods.

5. What are dunder methods in Python?

  - Dunder methods in Python are special methods with double underscores at the start and end of their names.
  - They let you define how objects behave with built-in functions or operators.

6. Explain the concept of inheritance in OOP?
  
  - Inheritance in OOP is when a new class gets properties and actions from an existing class.
  - The existing class is called the parent  class, and the new class is called the child class.
  - The child class can use everything from the parent class and also add its own features or change some behaviors.

7. What is polymorphism in OOP ?

 - Polymorphism in OOP means the same action can behave differently depending on the object.

- It lets you use the same method name for different classes, but each class can have its own way of doing that action.

12. What is method overriding in OOP?

  - Method overriding in OOP happens when a child class defines a method with the same name as a method in its parent class but gives it a new behavior.

  - It allows the child class to change or customize how the method works.

13. What is a property decorator in Python ?

  - This is useful when you want to control how data is accessed but still make it look simple.

15. What is an abstract class in Python?

17. What is the difference between a class variable and an instance variable?

 - Class Variable and Instance Variable are both used to store data in a class, but they are different in how they are shared.

 - Class Variable:
 - Shared by all objects of the class.
 - Defined outside of methods, usually at the top of the class.
 - Changing it affects all objects.

 - Instance Variable:
 - Unique to each object.
 - Defined inside the __init__ method using self.
 - Changing it affects only that object.

18. What is multiple inheritance in Python?

  - __repr__:

20. What is the significance of the ‘super()’ function in Python?

21. What is the significance of the __del__ method in Python?

24. What is method chaining in Python OOP?

25. What is the purpose of the __call__ method in Python?
